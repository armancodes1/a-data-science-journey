# Matplotlib
Matplotlib is a powerful library for creating static, animated, and interactive visualizations in Python.


It is widely used for data visualization in scientific computing.

## Getting Started

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Basic Line Plot

```python
x = [-2,-1,0,1,2]
y = [2,1,0,1,2]
plt.plot(x,y)
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.title('Sine Wave')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.show()
```

## Multiple Lines

```python
y2 = np.cos(x)
plt.plot(x, y, label='sin(x)')
plt.plot(x, y2, label='cos(x)')
plt.legend()
plt.title('Sine and Cosine')
plt.show()
```

## Scatter Plot

```python
x = np.random.rand(50)
y = np.random.rand(50)
sizes = 300*np.random.rand(50)
colors = np.random.rand(50)

plt.scatter(x, y, s=sizes, c=colors, alpha=0.5, cmap='viridis')
plt.title('Random Scatter Plot')
plt.colorbar()
plt.show()
```

## Bar Chart

```python
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 5, 2]

plt.bar(categories, values)
plt.title('Bar Chart Example')
plt.show()
```

## Histogram

```python
data = np.random.randn(1000)
plt.hist(data, bins=30, alpha=0.7, color='blue')
plt.title('Histogram of Random Data')
plt.show()
```

## Subplots

```python
x = np.linspace(0, 10, 100)

fig, axs = plt.subplots(2)
axs[0].plot(x, np.sin(x))
axs[0].set_title('Sine')

axs[1].plot(x, np.cos(x), 'r')
axs[1].set_title('Cosine')

plt.tight_layout()
plt.show()
```

## Heatmap using imshow

```python
matrix = np.random.rand(6,6)
plt.imshow(matrix, cmap='viridis', interpolation='nearest')
plt.colorbar()
plt.title('Heatmap Example')
plt.show()
```
