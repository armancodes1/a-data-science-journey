# SymPy

SymPy is a Python library for **symbolic mathematics**.

It allows algebraic manipulation, calculus, equation solving, and more.

## Getting Started

```python
import sympy as sp
sp.init_printing()
```

## Symbols
We define symbolic variables with `sp.symbols`. They behave like algebraic symbols.

```python
x, y, z = sp.symbols('x y z')
a, b = sp.symbols('a b', real=True)
expr = x + y + z
expr
```

## Algebraic Manipulation

```python
expr = (x + 2)*(x - 3)
expanded = sp.expand(expr)
factored = sp.factor(expanded)
expanded, factored
```

```python
expr = x**3 - 6*x**2 + 11*x - 6
sp.factor(expr)
```

### Simplification

```python
expr = sp.sin(x)**2 + sp.cos(x)**2
sp.simplify(expr)
```

### Substitution

```python
expr = x**2 + y
expr.subs(x, 3).subs(y, 4)
```

## Solving Equations

```python
eq = sp.Eq(x**2 - 2, 0)
sp.solve(eq, x)
```

```python
eq = sp.Eq(x**2 + 2*x + 1, 0)
sp.solve(eq, x)
```

### Systems of Equations

```python
eq1 = sp.Eq(2*x + y, 1)
eq2 = sp.Eq(x - y, 3)
sp.solve((eq1, eq2), (x,y))
```

## Calculus

### Derivatives

```python
f = sp.sin(x)*sp.exp(x)
sp.diff(f, x)
```

```python
sp.diff(x**4, x, 2)
```

### Integrals

```python
sp.integrate(sp.exp(-x**2), (x, -sp.oo, sp.oo))
```

```python
sp.integrate(sp.sin(x)/x, (x, 0, sp.oo))
```

### Limits

```python
sp.limit(sp.sin(x)/x, x, 0)
```

### Series Expansion

```python
sp.series(sp.cos(x), x, 0, 6)
```

## Linear Algebra

```python
M = sp.Matrix([[1,2],[3,4]])
M.det(), M.inv(), M.eigenvals()
```

## Differential Equations

```python
f = sp.Function('f')
deq = sp.Eq(f(x).diff(x,2) - 2*f(x).diff(x) + f(x), 0)
sp.dsolve(deq)
```

## Plotting

```python
sp.plot(sp.sin(x), (x, -2*sp.pi, 2*sp.pi))
```
