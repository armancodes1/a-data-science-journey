# Seaborn
---
Seaborn is a powerful Python library for **statistical data visualization** built on top of Matplotlib.

 It provides high-level functions for creating informative and attractive plots with minimal code.

## 1. Setup & Import

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load example dataset
sns.set(style='whitegrid')
tips = sns.load_dataset('tips')
tips.head()
```

## 2. Seaborn Themes & Styles
You can easily control the overall look using `sns.set_style()` and palettes.

```python
sns.set_style('darkgrid')
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title('Scatter with Darkgrid Style')
plt.show()
```

## 3. Basic Plots

```python
# Scatter Plot
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='sex')
plt.show()

# Line Plot
sns.lineplot(x='size', y='tip', data=tips, estimator='mean')
plt.show()

# Bar Plot
sns.barplot(x='day', y='total_bill', data=tips)
plt.show()
```

## 4. Distribution Plots

```python
# Histogram
sns.histplot(tips['total_bill'], bins=20, kde=True)
plt.show()

# Boxplot
sns.boxplot(x='day', y='total_bill', data=tips)
plt.show()

# Violin Plot
sns.violinplot(x='day', y='tip', data=tips, hue='sex', split=True)
plt.show()
```

## 5. Categorical Plots

```python
# Countplot
sns.countplot(x='day', data=tips)
plt.show()

# Stripplot
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True)
plt.show()

# Swarmplot
sns.swarmplot(x='day', y='total_bill', data=tips)
plt.show()
```

## 6. Relational Plots

```python
sns.relplot(x='total_bill', y='tip', hue='sex', col='time', data=tips)
plt.show()
```

## 7. Multi-variable Plots

```python
# Pairplot
sns.pairplot(tips, hue='sex')
plt.show()

# Heatmap
corr = tips.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
```

## 8. Customization
Seaborn integrates well with Matplotlib. You can set colors, titles, labels, and use palettes.

```python
sns.set_palette('pastel')
sns.barplot(x='day', y='total_bill', data=tips)
plt.title('Customized Barplot')
plt.xlabel('Day of Week')
plt.ylabel('Total Bill Amount')
plt.show()
```
