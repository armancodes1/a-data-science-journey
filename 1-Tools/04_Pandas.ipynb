# Pandas

Pandas is a Python library for **data analysis and manipulation**.

It provides two main structures: **Series** (1D) and **DataFrame** (2D).

## Getting Started

```python
import pandas as pd
import numpy as np
```

## Series

```python
s = pd.Series([10, 20, 30, 40], index=['a','b','c','d'])
s
```

## DataFrame

```python
data = {
    'Name': ['Alice','Bob','Charlie','Diana'],
    'Age': [25,30,35,40],
    'Salary': [50000,60000,70000,80000]
}
df = pd.DataFrame(data)
df
```

## Indexing and Selection

```python
df['Name']
```

```python
df[['Name','Salary']]
```

```python
df.iloc[0]
```

```python
df.loc[1,'Age']
```

## Filtering

```python
df[df['Age'] > 30]
```

## Adding Columns

```python
df['Bonus'] = df['Salary'] * 0.1
df
```

## Grouping and Aggregation

```python
data = {
    'Department': ['HR','HR','IT','IT','Finance','Finance'],
    'Employee': ['Alice','Bob','Charlie','Diana','Eve','Frank'],
    'Salary': [50000,55000,60000,65000,70000,72000]
}
df2 = pd.DataFrame(data)
df2.groupby('Department')['Salary'].mean()
```

## Sorting

```python
df.sort_values('Salary', ascending=False)
```

## Handling Missing Data

```python
df3 = pd.DataFrame({'A':[1,2,np.nan], 'B':[4,np.nan,6]})
df3
```

```python
df3.fillna(0)
```

```python
df3.dropna()
```

## Plotting

```python
import matplotlib.pyplot as plt

df['Salary'].plot(kind='bar', title='Salaries')
plt.show()
```
